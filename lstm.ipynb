```python
import pandas as pd
df2=pd.read_csv("test.txt",header=None)
c=df2.iloc[:,:-1]
c.head()
```

```python
df1=pd.read_csv("train.txt",header=None)
```

```python
x=df1.iloc[:,:-1]
y=df1.iloc[:,-1]
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
#from sk
from sklearn.model_selection import train_test_split
```

```python
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

def model():
    model=Sequential()
    model.add(Dense(500,activation="relu",input_shape=(499,1)))
    model.add(Dense(100,activation="relu"))
    model.add(Dense(10,activation="relu"))
    model.add(Dense(1,activation="relu"))
    
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model

estimator = KerasRegressor(build_fn=model, epochs=100, batch_size=5, verbose=0)
kfold = KFold(n_splits=10)
results = cross_val_score(estimator, x, y, cv=kfold)
```

```python
import xgboost as xgb
clf=xgb.XGBRegressor(random_state=40)
r=clf.fit(x,y)
train_set=clf.predict(c)

def modified_dataset(dataset,time):
    x_train=[]
    y_train=[]
    for i in range(len(dataset)-time-1):
        x_train.append(dataset[i:i+time])
        y_train.append(dataset[i+time])
    return (np.array(x_train),(y_train))    

x_train,y_train=modified_dataset(y,100)
train_set=np.array(train_set)
x_test,y_test=modified_dataset(train_set,100)
```

```python
x_train=x_train.reshape(x_train.shape[0],x_train.shape[1],1)
x_test=x_test.reshape(x_test.shape[0],x_test.shape[1],1)
x_test.shape

import keras 
from keras.layers import Dense,LSTM
from keras.models import Sequential 


model =Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(100,1)))
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))

model.compile(loss="mean_squared_error",optimizer="adam")

model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100,batch_size=32,verbose=1)
```
