```python
import pandas as pd
import numpy as np
from sklearn.metrics import silhouette_score,confusion_matrix,accuracy_score
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.metrics import calinski_harabasz_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.preprocessing import OneHotEncoder
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
data =  pd.read_csv(r"/content/data.csv")
data.set_index("id",inplace=True)
y = data.diagnosis.values.reshape(-1,1)
encode = OneHotEncoder(sparse=False)
y = encode.fit_transform(y)
data.drop(["Unnamed: 32","diagnosis"],axis=1,inplace=True)
```

```python
x = data
```

```python
## RMS:
def rms(data,i):
    return(np.sqrt(np.mean(np.square(data[i]))))
```

```python
def rms_gap(data,i):
    x = data[i].values
    gap=[abs(x[i+1] - x[i])  for i in range(1,len(data)-1)]
    gap_rms = np.sqrt(np.mean(np.square(gap)))
    temp = data[i] - min(data[i]) - gap_rms
    return(np.sqrt(np.mean(np.square(temp))))
```

```python
def rms_1(data,i):
    return(np.sqrt(np.mean(np.square( data[i]-min(data[i])))))
```

```python
def rms_mean(data,i):
    return(np.sqrt(np.mean(np.square( data[i]-np.mean(data[i])))))
```

```python
def gap_scale(data,i):
    return(max(data[i]) - min(data[i]))
```

####    Transformed Feature

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential,Model
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam,RMSprop
from tensorflow.keras.callbacks import EarlyStopping
from keras.initializers import RandomNormal,GlorotNormal
```

```python
score = {}
```

```python
def build_classifier():
  model = Sequential()
  model.add(Dense (8,input_shape=(30,),activation="relu"))
  model.add(Dense(32,activation="relu"))
  model.add(Dense(64,activation="relu"))
  model.add(Dense(2, activation="softmax"))
  model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
  return(model)
```

```python
classifier = KerasClassifier(build_fn = build_classifier, batch_size = 10, epochs = 100)
accuracies = cross_val_score(estimator = classifier, X = x, y = y, cv = 10)
```

```python
score[0]=[accuracies.min(),accuracies.mean(),accuracies.max()]
```
